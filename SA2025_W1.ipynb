### 1. What is name of car that has the highest horsepower?  

```python
max_hp_car = df[df['horsepower'] == df['horsepower'].max()]['name'].values
print("\nCar with highest horsepower:", max_hp_car)
```

### 2. How many cars have mpg  35?

```python
mpg_35_count = df[df['mpg'] >= 35].shape[0]
print("Number of cars with mpg  35:", mpg_35_count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
most_common_origin = filtered['origin'].mode()[0]
print("Most common origin (hp>100, weight<3000):", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
mean_acc_japan = round(df[df['origin'] == 'japan']['acceleration'].mean(), 2)
print("Mean acceleration of Japanese cars:", mean_acc_japan)
```

### 5. Which year had the highest average mpg?

```python
year_highest_avg_mpg = df.groupby('model_year')['mpg'].mean().idxmax()
print("Year with highest average mpg:", year_highest_avg_mpg)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = df['mpg'].median()
above_median = df[df['mpg'] > median_mpg]
best_ratio = above_median['hp_to_weight'].max()
best_ratio_cars = above_median[above_median['hp_to_weight'] == best_ratio][['name', 'hp_to_weight']]
print("\nBest hp-to-weight ratio among above-median mpg cars:")
print(best_ratio_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='model_year', y='mpg', hue='origin', estimator='mean')
plt.title("Average MPG Over Years by Origin")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),
    hue_order=['japan', 'europe', 'usa'],
    alpha=0.7
)
plt.title("Horsepower vs Weight (Colored by Origin, Sized by MPG)")
plt.xlabel("Horsepower")
plt.ylabel("Weight")
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc=2)
plt.show()

```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
name_counts = df['name'].value_counts()
multi_year_names = name_counts[name_counts > 1].index
consistent_cars = df[df['name'].isin(multi_year_names)]
grouped = consistent_cars.groupby('name')['mpg'].agg(['std', 'count', 'mean'])
consistent_filtered = grouped[(grouped['std'] < 1.0) & (grouped['count'] >= 2)]
consistent_sorted = consistent_filtered.sort_values(by=['count', 'mean'], ascending=False)

print("\nConsistent car models with low mpg variation:")
print(consistent_sorted.head(10))

```
